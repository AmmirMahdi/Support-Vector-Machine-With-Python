```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
```

## Load Data

```python
data = load_iris()
```

```python
# Convert To DataFrame
iris = pd.DataFrame(data.data, columns=data.feature_names)
iris['target'] = data.target
```

```python
iris.head()
```

## Create Model

```python
svc = SVC()
```

# Data => Train and Test

```python
from sklearn.cross_validation import train_test_split
```

```python
X = iris[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']]
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

### Set parameter for SVM
#### gamma =  [0.1, 0.01, 0.001, 0.0001]
#### C           =  [ 1, 10, 100, 1000 ]
#### kernel = Linear

```python
param_grid = {'C':(1, 10, 100, 1000), 'gamma':(0.1, 0.01, 0.001, 0.0001)}
```

```python
param_grid
```

```python
grid_search = GridSearchCV(svc, param_grid, cv=10, verbose=10)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
svc = SVC(C=10, gamma=0.1).fit(X_train, y_train)
```

```python
svc.score(X_test, y_test)
```

```python
y_pred = svc.predict(X_test)
```

```python
print("Confusion Matrix: ")
print(confusion_matrix(y_test, y_pred))
print()
print("Classifiction Report :")
print(classification_report(y_test, y_pred))
```
